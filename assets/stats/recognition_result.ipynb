```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt

# Specify the directory path where the JSON files are located
directory = "../model_result/"

# Get all the names from the directory
names = os.listdir(directory)

# Print the names as a list of strings
print(names)

dlib_path_list = ['dlib_test_base.json', 'dlib_test_nolook.json', 'dlib_test_specs.json', 'dlib_test_mask.json', 'dlib_test_hat_specs.json', 'dlib_test_3ppl.json', 'dlib_test_dim.json']

facenet_path_list = ['facenet_test_base.json', 'facenet_test_nolook.json', 'facenet_test_specs.json', 'facenet_test_mask.json', 'facenet_test_hat_specs.json', 
                'facenet_test_3ppl.json', 'facenet_test_dim.json']
```

```python
def load_data(directory, path_str):

        # Create the file path
        file_path = os.path.join(directory, path_str)

        # Check if the file has a JSON extension
        if file_path.endswith(".json"):
            # Open the JSON file and load the data
            with open(file_path, "r") as file:
                data = json.load(file)

                # print(data)

                # Extract the details from the loaded data
                model_name = data['Model Name']
                dataset_name = data['Dataset Name']
                total_time_spent = data['Total Time Spent']
                correct = data['Correct']
                total_images = data['Total Images']
                total_accuracy = data['Total Accuracy']
                total_precision = data['Total Precision']
                total_recall = data['Total Recall']
                total_f1_score = data['Total F1 Score']

        return [model_name, dataset_name, total_time_spent, correct, total_images, total_accuracy, total_precision, total_recall, total_f1_score]

```

```python
for idx in range(7):

    dlib_path = dlib_path_list[idx]
    facenet_path = facenet_path_list[idx]

    # Initialize variables to store the extracted details
    dlib = load_data(directory, dlib_path)
    facenet = load_data(directory, facenet_path)


    # Plot the barchart
    result = pd.DataFrame(
        {
            'Dlib': dlib[-4:],
            'Facenet': facenet[-4:],
        },
        index=['Accuracy', 'Precision', 'Recall', 'F1-Score'])
    result.plot(kind='bar', rot=0, figsize=(15, 10), title=f'Recognition Model Evaluation for {dlib[1]} dataset')
    plt.xlabel('Recognition Model Evaluation Metrics')
    plt.ylabel('Values')
    plt.show()
    display(result)
```

```python
dlib_pre = []
facenet_pre = []

for idx in range(7):

    dlib_path = dlib_path_list[idx]
    facenet_path = facenet_path_list[idx]

    # Initialize variables to store the extracted details
    dlib_pre.append(load_data(directory, dlib_path)[-3])
    facenet_pre.append(load_data(directory, facenet_path)[-3])

    
# Plot the barchart
result = pd.DataFrame(
    {
        'Dlib': dlib_pre,
        'Facenet': facenet_pre,
    },
    index=['test_base', 'test_nolook', 'test_specs', 'test_mask', 'test_hat_specs', 'test_3ppl', 'test_dim'])
result.plot(kind='bar', rot=0, figsize=(15, 10), title=f'Overall Precision of Recognition Model in different dataset')
plt.xlabel('Recognition Model Evaluation Datasets')
plt.ylabel('Values')
plt.show()
display(result)
```
