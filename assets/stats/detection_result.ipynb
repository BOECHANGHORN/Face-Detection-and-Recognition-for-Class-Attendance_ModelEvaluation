```python
import pandas as pd
import matplotlib.pyplot as plt

path_str_list = ['test_base', 'test_nolook', 'test_specs', 'test_mask', 'test_hat_specs', 'test_3ppl', 'test_dim']


total_face = [[218, 109, 126], [177, 99, 127], [94, 87, 118], [38, 6, 112], [95, 24, 109], [397, 300, 374], [357, 191, 246]]
tp = [[108, 109, 122], [77, 85, 115], [89, 87, 114], [0, 6, 104], [0, 21, 53], [291, 292, 338], [215, 187, 228]]
```

```python
for i in range(len(path_str_list)):

    path_str = path_str_list[i]

    total_faces_detected = total_face[i]
    TP = tp[i]

    # Calculate metric
    accuracy = [TP[i] / max(total_faces_detected[i], 1) for i in range(len(total_faces_detected))]

    haar_cascade = [accuracy[0]]
    hog = [accuracy[1]]
    mtcnn = [accuracy[2]]

    # Plot the barchart
    result = pd.DataFrame(
        {
            'Haar Cascade': haar_cascade,
            'HOG': hog,
            'MTCNN': mtcnn
        },
        index=['Accuracy'])
    result.plot(kind='bar', rot=0, figsize=(15, 10), title=f'Detection Model Evaluation for {path_str} dataset')
    plt.xlabel('Detection Model Evaluation Metrics')
    plt.ylabel('Values')
    plt.show()
    display(result)
```
